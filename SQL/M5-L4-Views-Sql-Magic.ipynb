````python
%%sql

SELECT * 
FROM survey s 
    JOIN visited v ON (v.id::int = s.taken::int) 
    JOIN person p ON (s.person=p.id)
````

## Clarity through Column Renames 
One of the issues we see is that the `id` column exists twice.
First from the `visited.id` column, then from the `person.id` column.

We need to fix this to have unambiguous columns.

We also see that the `person` column of the output is redundant with the `person.id` column. 

Let's update our SQL to make the results more clear.

````python
%%sql

SELECT v.id as visit_id, 
    s.quant, s.reading,
    v.site, v.dated as visit_date,
    p.id as person_id, p.personal as firstname, p.family as lastname
FROM survey s 
JOIN visited v ON (v.id::int = s.taken::int) 
JOIN person p ON (s.person=p.id);
````

## Now that we have this _report_ from the database, let's make this accessible for future use without re-writing the SQL.

To do this, we will create a view, wich is basically a stored, named `table expression`. Recalling a select statement is the definition of a table, a table expression, then:  
`CREATE OR REPLACE VIEW SSO.NAME AS SELECT_STATEMENT`

````python
%%sql

CREATE OR REPLACE VIEW survey_view  -- clause for creating a view
AS
SELECT v.id as visit_id, 
    s.quant, s.reading,
    v.site, v.dated as visit_date,
    p.id as person_id, p.personal as firstname, p.family as lastname
FROM survey s 
JOIN visited v ON (v.id::int = s.taken::int) 
JOIN person p ON (s.person=p.id);
````

## Selecting data from a view
#### ... is the same as selecting data from single table.

````python
%%sql

SELECT * FROM survey_view
````

### Is it possible to check the definition of a table using a query?

It is possible to get table descripiton via writing a query. E.g., `information_schema.views` catalog contains view definitions and `information_schema.columns` catalog contains the information on columns of all tables/views. We can probe these table to see whether view is created or not. 



````python
%%sql

-- select * 
select table_schema, table_name, view_definition
from information_schema.views
where table_name = 'survey_view'

````

````python
%%sql

-- SELECT * 
SELECT table_name, column_name, data_type, column_default, is_nullable 
FROM 
   information_schema.columns
WHERE 
   table_name = 'survey_view';
    
````
